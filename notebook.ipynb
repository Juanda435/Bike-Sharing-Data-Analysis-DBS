memprediksi jumlah peminjaman sepeda

```python
# Misalkan df sudah memiliki data dengan kolom 'dteday' dan 'cnt'
df = day_df.copy()

# Pastikan data diurutkan berdasarkan tanggal
df = df.sort_values(by="dteday")

# Tambahkan kolom Moving Average (7 hari terakhir)
df["moving_avg_7"] = df["cnt"].rolling(window=7).mean()

# Prediksi jumlah peminjaman untuk hari berikutnya (hari ke-31)
predicted_rental = df["moving_avg_7"].iloc[-1]

print(f"{predicted_rental:.0f}")
```

membuat sebuah DataFrame baru bernama all_df untuk menampung semua informasi dari kedua tabel tersebut

```python
# Gabungkan day_df dengan hours_df berdasarkan tanggal (dteday)
all_df = hours_df.merge(day_df, on="dteday", suffixes=("_hour", "_day"))

# Tampilkan 5 baris pertama untuk memastikan penggabungan sukses
all_df.head()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:
- Bagaimana pengaruh cuaca terhadap jumlah peminjaman sepeda?

```python
# Kelompokkan data berdasarkan kondisi cuaca dan hitung rata-rata penyewaan sepeda
weather_rentals = all_df.groupby("weathersit_day")["cnt_hour"].mean().reset_index()

# Buat plot
plt.figure(figsize=(8,5))
sns.barplot(data=weather_rentals, x="weathersit_day", y="cnt_hour", hue="weathersit_day", palette="viridis", legend=False)

# Tambahkan judul dan label
plt.title("Pengaruh Cuaca terhadap Peminjaman Sepeda", fontsize=14)
plt.xlabel("Kondisi Cuaca", fontsize=12)
plt.ylabel("Rata-rata Jumlah Peminjaman per Jam", fontsize=12)
plt.xticks(ticks=[0, 1, 2, 3], labels=["Cerah", "Mendung", "Hujan Ringan", "Hujan Lebat"])
plt.show()
```

insight:
- peminjaman paling banyak terjadi pada cuaca yang cerah hal ini menunjukkan
  bahwa kondisi cuaca yang nyaman mendorong lebih banyak orang untuk menyewa sepeda.
- Saat cuaca mendung, peminjaman sedikit menurun dibandingkan saat cerah,  
  tetapi masih cukup tinggi. Kemungkinan karena meskipun mendung, kondisi masih nyaman untuk bersepeda.
- Saat hujan ringan, jumlah peminjaman sepeda mengalami penurunan yang cukup
  drastis. Ini menunjukkan bahwa pengguna cenderung menghindari bersepeda saat ada hujan.
- Pada kondisi hujan lebat, jumlah peminjaman hampir nol. Ini masuk akal karena
  kondisi tersebut tidak nyaman dan bisa berbahaya bagi pesepeda.


### Pertanyaan 2:
- Pada jam berapa peminjaman sepeda mencapai puncaknya dalam satu hari?

```python
# Kelompokkan data berdasarkan jam dan hitung rata-rata peminjaman sepeda
hourly_rentals = all_df.groupby("hr")["cnt_hour"].mean()

# Cari jam dengan peminjaman tertinggi
peak_hour = hourly_rentals.idxmax()
max_rentals = hourly_rentals.max()

# Buat visualisasi
plt.figure(figsize=(10,5))
sns.lineplot(x=hourly_rentals.index, y=hourly_rentals.values, marker="o", color="b")

# Tambahkan garis vertikal untuk jam puncak
plt.axvline(x=peak_hour, color="r", linestyle="--", label=f"Peak Hour: {peak_hour}:00 ({int(max_rentals)} rentals)")

# Tambahkan judul dan label
plt.title("Tren Peminjaman Sepeda Berdasarkan Jam", fontsize=14)
plt.xlabel("Jam", fontsize=12)
plt.ylabel("Rata-rata Jumlah Peminjaman", fontsize=12)
plt.xticks(range(0, 24))
plt.legend()
plt.show()
```

**Insight:**
- Puncak peminjaman ada di jam 17:00,Ini kemungkinan besar karena jam pulang
  kerja atau pulang sekolah.
- Ada lonjakan juga di pagi hari sekitar jam 7-8,Ini bisa jadi jam orang
  berangkat kerja atau sekolah.
- Siang hari relatif stabil, tapi lebih rendah,Antara jam 10-15, jumlah
  peminjaman cenderung rata tapi gak sebanyak pagi atau sore. Mungkin ini karena orang udah di tempat kerja/sekolah dan belum pulang.
- Malam hari peminjaman turun drastis,Setelah jam 18:00, jumlah peminjam mulai
  turun terus sampai tengah malam,kemungkinan karena tidak ada aktivitas lagi

melihat pada holiday jam berapa biasanya orang menyewa

```python
# Filter hanya untuk hari libur
holiday_df = all_df[all_df["holiday_hour"] == 1]

# Hitung rata-rata peminjaman per jam
holiday_rentals = holiday_df.groupby("hr")["cnt_hour"].mean()

# Plot hasilnya
plt.figure(figsize=(10, 5))
sns.lineplot(x=holiday_rentals.index, y=holiday_rentals.values, marker="o", color="b")

# Tambahkan garis puncak jam sibuk
peak_hour = holiday_rentals.idxmax()
peak_rentals = holiday_rentals.max()
plt.axvline(peak_hour, color="red", linestyle="--", label=f"Peak Hour: {peak_hour}:00 ({int(peak_rentals)} rentals)")

# Atur label & judul
plt.xlabel("Jam")
plt.ylabel("Rata-rata Jumlah Peminjaman")
plt.title("Tren Peminjaman Sepeda pada Hari Libur")
plt.legend()
plt.grid(True)

# Tampilkan plot
plt.show()
```

insight
- pada hari libur orang lebih sering penyewa sepeda pada jam siang sekitar jam
